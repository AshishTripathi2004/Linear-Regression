```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
## load the dataset
df = pd.read_csv('insurance.csv')
df.head()
```

```python
## data shape
df.shape
```

```python
## data details
df.info()
```

```python
# check for nulls
df.isna()
```

```python
## description of data
df.describe()
```

```python
# create pairplots for data
sns.jointplot(x='age',y='charges',data=df)
plt.show()
```

```python
# Encode 'sex': male=1, female=0
df['sex_enc'] = df['sex'].map({'male': 1, 'female': 0})

# Encode 'smoker': yes=1, no=0
df['smoker_enc'] = df['smoker'].map({'yes': 1, 'no': 0})
```

```python
# create all pairwise plots
sns.pairplot(df, kind='scatter',plot_kws={'alpha':0.5})
plt.show()
```
