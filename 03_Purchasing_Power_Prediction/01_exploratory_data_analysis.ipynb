```python
# load the dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# read the dataset from the csv file
df=pd.read_csv('customer.csv')
df.head()
```

```python
#analyse the dataset characteristics
print("The dataset shape : ", df.shape)
print("Number of rows : ", df.shape[0])
print("Number of features : ", df.shape[1])
```

```python
#get the columns list and data types
print("The dataset information : ")
print(df.info())
```

```python
#analyse the dataset characteristics and statistics
print(df.describe())
```

```python
# create pair plots
sns.jointplot(x='Time on Website', y='Yearly Amount Spent',data=df,color='skyblue',alpha=0.5)
sns.jointplot(x='Time on App', y='Yearly Amount Spent',data=df,color='green',alpha=0.5)
sns.jointplot(x='Length of Membership', y='Yearly Amount Spent',data=df,color='pink',alpha=0.5)
sns.jointplot(x='Avg. Session Length', y='Yearly Amount Spent',data=df,color='red',alpha=0.5)
plt.show()
```

```python
# create all pairplots all at once
sns.pairplot(df, kind='scatter',plot_kws={'alpha':0.5})
```

```python
sns.lmplot(x='Length of Membership',y='Yearly Amount Spent',data=df,scatter_kws={'alpha':0.4})
plt.show()
```
