## Linear Regression Implementation

```python
## import pandas for loading the dataset and matplotlib for data visualisation
import pandas as pd
import matplotlib.pyplot as plt
```

```python
## load the dataset
data = pd.read_csv('../01_Linear_Regression_Fundamentals/study_time.csv')
data.head()
```

```python
## Create a scatter plot
x = data['StudyTime']
y = data['ExamScore']

plt.scatter(x,y,color='blue')
plt.show()
```

```python
## define the loss function
## generic to the linear regression problem
## never generally used
## instead we use the gradient descent optimiser

def loss_function(m,b):

    x=data['StudyTime']
    y=data['ExamScore']

    loss=0
    for i in range(len(data)):
        loss+=(y.iloc[i]-(m*x.iloc[i]+b))**2
    
    loss=loss/(float(len(data)))
    return loss
```

```python
## gradient descent optimiser
def gradient_descent_optimiser(m,b,L):

    x=data['StudyTime']
    y=data['ExamScore']

    m_gradient=0
    b_gradient=0
    for i in range(len(data)):
        error = (y.iloc[i]-(x.iloc[i]*m+b))

        m_gradient+=(x.iloc[i]*error)
        b_gradient+=(error)
    
    m_gradient = -(2*m_gradient)/(float(len(data)))
    b_gradient = -(2*b_gradient)/(float(len(data)))

    m_new = m - L*m_gradient
    b_new = b - L*b_gradient

    return m_new, b_new
```

```python
## find out the value of m and b for best fit line
epochs=1000
m=0
b=0
L=0.0001
for i in range(epochs):
    if(i%50==0):
        print('Epoch',i)
    m,b=gradient_descent_optimiser(m,b,L)
```

```python
## plot the best fit line
## Create a scatter plot
x = data['StudyTime']
y = data['ExamScore']

plt.scatter(x,y,color='blue')
plt.plot(x,[m*x+b for x in x],color='red')
plt.show()
```
